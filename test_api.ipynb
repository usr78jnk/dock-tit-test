С этого я получил сл. знания как можно обращатся к сервисам в интернете:
1. Обычным `curl` - запросом: `curl -X GET http://127.0.0.1:5000/health`. Что подходит для быстрой роверки ресурса.
2. Через браузер Web interface `http://127.0.0.1:5000/docs` - что тоже возможно, сразу все просмотреть.
3. Ну и через написание своего **request** - запроса

## Test API

```python
import requests

def predict(data):
    url = "http://127.0.0.1:5000/predict"
    response = requests.post(url=url, json=data)

    if response.status_code == 200:
        return response.json()
    else: 
        return f"Error opening page:{response.status_code}"

data = {
    "Pclass":7,
    "Age": 30,
    "Fare": 40
}
# Попробую уже сам написать get хапрос на получение к-ва запросов
def count():
    url = "http://127.0.0.1:5000/stats"
    response = requests.get(url=url)
    if response.status_code == 200:
        return response.json()
    else:
        return f"Error opening page:{response.status_code}"



# получение и возврат результатов
resulte = predict(data)
print(resulte)
print(count())
```
